```python
#범위 내부의 정수를 모두 더하는 함수
def sum_all(start, end):
    output = 0
    for i in range(start, end + 1):
        output += i

    return output

print("0 to 100:", sum_all(0, 100))
print("0 to 1000:", sum_all(0, 1000))
print("50 to 100:", sum_all(50, 100))
print("500 to 1000:", sum_all(500, 1000))
```

```python
def sum_all(start=0, end=100, step=1):
    output = 0
    for i in range(start, end+1, step):
        output += i
    return output

print("A.", sum_all(0, 100, 10))
print("B.", sum_all(end=100))
print("C.", sum_all(end=100, step=2))
```

```python
def mul(*values):
    output = 1
    for value in values:
        output *= value
    return output
    
print(mul(5,7,9,10))
```

```python
def factorial(n):
    output = 1
    for i in range(1, n+1):
        output *= i
    
    return output

print("1!:", factorial(1))
print("5!:", factorial(5))
```

```python
#재귀 함수 사용하기
def factorial(n):
    if n ==0:
        return 1
    else:
        return n * factorial(n-1)

print("6!:", factorial(6))
```

* 재귀함수는 상황에 따라서 기하급수 적으로 많이 반복한다는 문제가 있음

```python
# 메모화, memorization
#피보나치 수열 구현하기

def fibonacci(n):
    if n == 1:
        return 1
    if n == 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

print("fibonacci(5):", fibonacci(5))
print("fibonacci(7):", fibonacci(7))
print("fibonacci(30):", fibonacci(30))
```

```python
counter = 0 # 변수 선언

def fibonacci(n):
    # 어떤 피보나치 수를 구하는지 출력
    print("fibonacci({})를 구합니다.".format(n))
    global counter # 함수 내에 있는 변수를 전역변수로 사용하겠다.
    counter += 1

    if n == 1:
        return 1
    if n == 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

fibonacci(10)
print("---")
print("fibonacci(10) 계산에 활용된 덧셈 횟수는 {}번입니다.".format(counter))
```

* 변수에 접근하는 것을 참조
global 키워드 : 파이썬 프로그래밍 언어에만 있는 특이한 구조로, 
'W0621: Redefining name 'counter' from outer scope 또는 E0602: Undefined variable 'counter''가 나타나면 global 키워드를 쓰라는 것으로 판단, 수정

```python
# memorization
dictionary = {
    1: 1,
    2: 1
}

def fibonacci(n):
    if n in dictionary: #메모가 되어 있으면 메모된 값을 리턴
        return dictionary[n]
    else:
        # 메모가 되어 있지 않으면 값을 구함
        output = fibonacci(n-1) + fibonacci(n-2)
        dictionary[n] = output
        return output

print("fibonacci(10):", fibonacci(10))
print("fibonacci(50)", fibonacci(50))
```

```python
def flatten(data):
    output = []
    for item in data:
        if type(item) == list:
            output += flatten(item) # type이 list형이면 output에 더하기
        else:
            output.append(item) #list형이 아니면 바로 output에 원소로 집어넣기
    return output

example = [[1,2,3], [4,[5,6]], 7, [8,9]]
print("원본:", example)
print("변환:", flatten(example))
```

```python
#테이블에 나눠 앉기(분할 문제)
# 한사람만 앉는 것은 안됨

minimum = 2
maximum = 10
total = 100
memo = {} # 딕셔너리 생성

def problem(rest, present):
    key = str([rest, present])
    #종료 조건
    if key in memo:
        return memo[key]
    if rest < 0:
        return 0 # 무효하니 0을 리턴
    if rest == 0:
        return 1 # 유효하니 수를 세면 되서 1을 리턴
    #재귀 처리
    count = 0 
    for i in range(present, maximum +1):
        count += problem(rest-i, i)
    #메모화 처리
    memo[key] = count
    #종료
    return count

print(problem(total, minimum))
```

```python
#테이블에 나눠 앉기
# 한사람만 앉는 것은 안됨

minimum = 2
maximum = 10
total = 100
memo = {}

def problem(rest, present):
    key = str([rest, present])
    #종료 조건
    if key in memo:  #key값이 memo 안에 있으면 memo[key]값 리턴
        return memo[key]
    if rest < 0:
        return 0 # 무효이므로 0을 리턴
    if rest == 0:
        return 1 # 유효하므로 수를 세기위해 1을 리턴
    #재귀 처리
    count = 0
    for i in range(present, maximum +1): # 앉힌 사람수부터 앉힐 수 있는 사람수
        count += problem(rest-i, i) # 남아있는 사람 수에 앉힌 사람수를 뺀 값, 앉힌 사람수를 재귀
    #메모화 처리
    memo[key] = count
    #종료
    return count

print(problem(total, minimum))
```

* map(함수, 리스트) -> 리스트의 요소를 함수에 넣고 리턴된 값으로 새로운 리스트를 구성
* filter(함수, 리스트) -> 리스트의 요소를 함수에 넣고 리턴된 값이 True인 것. 새로운 리스트 구성

```python
def power(item):
    return item * item

def under_3(item):
    return item < 3

#변수를 선언합니다.
list_input_a = [1,2,3,4,5]

# map() 함수 사용
output_a = map(power, list_input_a)
print("# map 함수의 실행결과")
print("map(power, list_input_a):", output_a)
print("map(power, list_input_a):", list(output_a))
print()

# filter 함수를 사용하기
output_b = filter(under_3, list_input_a)
print("#filter 함수의 실행결과")
print("filter(under_3, list_input_a):", output_b)
print("filter(under_3, list_input_a):", list(output_b))
```

* CSV = Comma Separated Values, 쉼표로 구분된 값

```python
import random
hanguls = list("가나다라마바사아자차카타파하")

with open("info.txt", "w") as file:
    for i in range(1000):
        name = random.choice(hanguls) + random.choice(hanguls)
        weight = random.randrange(40, 100)
        height = random.randrange(140, 200)

        file.write("{}, {}, {}\n".format(name,weight,height))
```

* for 한 줄을 나타내는 문자열 in 파일 객체:
    <br/>   처리

```python
with open("info.txt", "r") as file:
    for line in file:
        (name, weight, height) = line.strip().split(",")

        if (not name) or (not weight) or (not height):
            continue
        
        bmi = int(weight) / ((int(height)/ 100)**2)
        result = ""
        if 25 <= bmi:
            result = "과체중"
        elif 18.5 <= bmi:
            result = "정상 체중"
        else:
            result = "저체중"

        print('\n'.join([
                         "이름: {}",
                         "몸무게: {}",
                         "키: {}",
                         "BMI: {}",
                         "결과: {}"

        ]).format(name, weight, height, bmi, result))
        print()
```
